```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat Color pallete untuk mempercantik tampilan
colors_green = ["#004D00", "#197419", '#4CAF50', '#80C080', '#C0E0C0']
colors_mix = ["#17869E", '#264D58', '#179E66', '#D35151', '#E9DAB4', '#E9B4B4', '#D3B651', '#6351D3']
colors_div = ["#004D00", '#4CAF50', '#DADADA', '#D35151', '#331313']

sns.palplot(colors_green)
plt.title('Green Palette')
plt.show()

sns.palplot(colors_mix)
plt.title('Mixed Palette')
plt.show()

sns.palplot(colors_div)
plt.title('Diverse Palette')
plt.show()

# Ubah warna menggunakan palet yang telah dibuat
sns.set_palette(colors_green)

# Filter rows for Indonesia and Iraq
countries_third = ['Indonesia', 'Iraq']
df_plot_third = df_third_countries[df_third_countries['Country'].isin(countries_third)]

# Columns to include in the radar chart
columns = ['Dystopia (1.83) + residual', 'Explained by: GDP per capita',
           'Explained by: Social support', 'Explained by: Healthy life expectancy',
           'Explained by: Freedom to make life choices',
           'Explained by: Generosity', 'Explained by: Perceptions of corruption']

# Number of variables we're plotting.
num_vars = len(columns)

# Compute angle each bar is centered on:
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # Complete the loop by appending the first angle at the end.

# Draw the radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

for k, (idx, row) in enumerate(df_plot_third.iterrows()):
    data = row[columns].tolist() + row[columns].tolist()[:1]  # Complete the loop
    ax.fill(angles, data, alpha=0.4, label=row['Country'], color=colors_green[3*k], edgecolor=colors_green[3*k])
    ax.plot(angles, data)

# Labels for each point
ax.set_xticks(angles[:-1])
col_name = ['Dystopia', 'GDP per capita',
           'Social support', 'Healthy life expectancy',
           'Freedom to make life choices',
           'Generosity', 'Perceptions of corruption']
ax.set_xticklabels(col_name, fontsize=10)  # Menjadikan tulisan di sumbu x lebih jelas

# Title and legend
plt.title('Comparison of Happiness Factors: Iraq vs. Indonesia', size=15, color=colors_green[1], y=1.1)
ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))
plt.show()


# Filter rows for Indonesia and Iraq
countries_third = ['Indonesia', 'Iraq']
df_plot_third = df_third_countries[df_third_countries['Country'].isin(countries_third)]

# Columns to include in the stacked bar chart
columns = ['Dystopia (1.83) + residual', 'Explained by: GDP per capita',
           'Explained by: Social support', 'Explained by: Healthy life expectancy',
           'Explained by: Freedom to make life choices',
           'Explained by: Generosity', 'Explained by: Perceptions of corruption']

# Set data for each country
category_indo = df_plot_third[df_plot_third['Country'] == 'Indonesia'][columns].values.flatten()
category_iraq = df_plot_third[df_plot_third['Country'] == 'Iraq'][columns].values.flatten()

# Set the labels for each category
labels = columns

# Set the positions for the bars
x = np.arange(len(labels))

# Set the width of the bars
width = 0.35

# Plot the stacked bars
fig, ax = plt.subplots(figsize=(12, 8))

# Plot bars for Indonesia
bars_indo = ax.bar(x - width/2, category_indo, width, label='Indonesia', color='green')

# Plot bars for Iraq on top of Indonesia's bars
bars_iraq = ax.bar(x + width/2, category_iraq, width, label='Iraq', color='lightcoral')

# Add labels to the bars (Indonesia)
for i, (category, value) in enumerate(zip(labels, category_indo)):
    ax.text(i - width/2, value, f'{value:.2f}', ha='center', va='bottom', color='black')

# Add labels to the bars (Iraq)
for i, (category, value) in enumerate(zip(labels, category_iraq)):
    ax.text(i + width/2, value, f'{value:.2f}', ha='center', va='bottom', color='black')

# Add labels, title, and legend
ax.set_ylabel('Explained by Score')
ax.set_xlabel('Factors')
ax.set_title('Comparison of Happiness Factors: Iraq vs. Indonesia')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Adjust layout
plt.tight_layout()
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)

# Show plot
plt.show()


# Perbandingan antara Japan dan Uzbekistan

# Filter rows for Japan and Uzbekistan
countries_final = ['Japan', 'Uzbekistan']
df_plot_final = df_final_countries[df_final_countries['Country'].isin(countries_final)]

# Columns to include in the radar chart
columns = ['Dystopia (1.83) + residual', 'Explained by: GDP per capita',
           'Explained by: Social support', 'Explained by: Healthy life expectancy',
           'Explained by: Freedom to make life choices',
           'Explained by: Generosity', 'Explained by: Perceptions of corruption']

# Number of variables we're plotting.
num_vars = len(columns)

# Compute angle each bar is centered on:
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # Complete the loop by appending the first angle at the end.

# Draw the radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

for k, (idx, row) in enumerate(df_plot_final.iterrows()):
    data = row[columns].tolist() + row[columns].tolist()[:1]  # Complete the loop
    ax.fill(angles, data, alpha=0.4, label=row['Country'], color=colors_green[3*k], edgecolor=colors_green[3*k])
    ax.plot(angles, data)

# Labels for each point
ax.set_xticks(angles[:-1])
col_name = ['Dystopia', 'GDP per capita',
           'Social support', 'Healthy life expectancy',
           'Freedom to make life choices',
           'Generosity', 'Perceptions of corruption']

# Set labels for x-axis with rotation
ax.set_xticklabels(col_name, rotation=45, ha='right', fontsize=10)

# Title and legend
plt.title('Comparison of Happiness Factors: Iraq vs. Indonesia', size=15, color=colors_green[1], y=1.1)
ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))

# Adjust layout
plt.tight_layout()
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.3)

# Filter rows for Japan and Uzbekistan
countries_final = ['Japan', 'Uzbekistan']
df_plot_final = df_final_countries[df_final_countries['Country'].isin(countries_final)]

# Columns to include in the stacked bar chart
columns = ['Dystopia (1.83) + residual', 'Explained by: GDP per capita',
           'Explained by: Social support', 'Explained by: Healthy life expectancy',
           'Explained by: Freedom to make life choices',
           'Explained by: Generosity', 'Explained by: Perceptions of corruption']

# Set data for each country
category_japan = df_plot_final[df_plot_final['Country'] == 'Japan'][columns].values.flatten()
category_uzbek = df_plot_final[df_plot_final['Country'] == 'Uzbekistan'][columns].values.flatten()

# Set the labels for each category
labels = columns

# Set the positions for the bars
x = np.arange(len(labels))

# Set the width of the bars
width = 0.35

# Plot the stacked bars
fig, ax = plt.subplots(figsize=(12, 8))

# Plot bars for Japan
bars_japan = ax.bar(x - width/2, category_japan, width, label='Japan', color='green')

# Plot bars for Uzbekistan on top of Japan's bars
bars_uzbek = ax.bar(x + width/2, category_uzbek, width, label='Uzbekistan', color='skyblue')

# Add labels to the bars (Japan)
for i, (category, value) in enumerate(zip(labels, category_japan)):
    ax.text(i - width/2, value, f'{value:.2f}', ha='center', va='bottom', color='black')

# Add labels to the bars (Uzbekistan)
for i, (category, value) in enumerate(zip(labels, category_uzbek)):
    ax.text(i + width/2, value, f'{value:.2f}', ha='center', va='bottom', color='black')

# Add labels, title, and legend
ax.set_ylabel('Explained by Score')
ax.set_xlabel('Factors')
ax.set_title('Comparison of Happiness Factors: Japan vs. Uzbekistan')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Adjust layout
plt.tight_layout()
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)


# Show plot
plt.show()
```
